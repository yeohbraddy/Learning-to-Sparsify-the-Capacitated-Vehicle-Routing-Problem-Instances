```python
import numpy as np
import matplotlib.pyplot as plt
import re
from linecache import getline
from gurobipy import Model, GRB, quicksum
import pandas as pd
```

```python
def parseCoords(line):
    x = line.split()
    return [int(x[0]), int(x[1]), int(x[2])]
```

```python
COORD_FLAG = False
xc = []
yc = []
coords = {}
    
fh = open("A-n32-k5.txt",'r')
for i,line in enumerate(fh):
    if "NODE_COORD_SECTION" in line:
        COORD_FLAG = True
    elif "DEMAND_SECTION" in line:
        break
    elif COORD_FLAG:
        coord = parseCoords(line)
        xc.append(coord[1])
        yc.append(coord[2])
        coords[coord[0]] = [coord[1], coord[2]]
fh.close()
```

```python
print(coords)
```

```python
def parseRoutes(line):
    x = line.split()
    return x[2:]
```

```python
fh = open("test.txt",'r')  # A-n32-k5-solution.txt
routes = {}
for i, line in enumerate(fh):
    if "Route #" in line:
        routes[i] = parseRoutes(line)
        routes[i] = [ int(x) for x in routes[i] ]
fh.close()
print(routes)
```

```python
plt.rcParams["figure.figsize"] = [16,9]

for route in routes:
    currRoute = routes[route]
    node0 = currRoute[0]
    coord0 = coords[node0]
    plt.plot([xc[0],  coord0[0]], [yc[0], coord0[1]], c='g', zorder=0)
    for i in range(0, len(currRoute) - 1):
        node1 = currRoute[i + 1]
        node0 = currRoute[i]
        coord0 = coords[node0]
        coord1 = coords[node1]
        plt.plot([coord0[0], coord1[0]], [coord0[1], coord1[1]], c='g', zorder=0)
        
    plt.plot([xc[0], coord1[0]], [yc[0], coord1[1]], c='g', zorder=0)


plt.plot(xc[0], yc[0], c='r', marker='s')
plt.scatter(xc[1:], yc[1:], c='b')
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=16)
```

```python
from collections import defaultdict
data = defaultdict(list)

for coord in coords:
    data["Node_ID"].append(coord)
    data["X_Coord"].append(coords[coord][0])
    data["Y_Coord"].append(coords[coord][1])
    
    for route in routes:
        if coord in routes[route]:
            data["Route_ID"].append(route)
            data["IsOptimal"].append(1)
        else:
            data["Route_ID"].append(-1)
            data["IsOptimal"].append(0)
        break
        
for key in data:
    print(key, data[key], len(data[key]))
```
